```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pathToData_All = 'C:/Users/Dmitriy/Desktop/Job/CoursData/data/processed/COALL.csv'
pathToData_L = 'C:/Users/Dmitriy/Desktop/Job/CoursData/data/processed/COL.csv'
pathToData_M = 'C:/Users/Dmitriy/Desktop/Job/CoursData/data/processed/COM.csv'
```

```python
data_L = pd.read_csv(pathToData_L)
data_M = pd.read_csv(pathToData_M)
```

```python
data_L.drop([0,1,2], inplace=True)
data_M.drop([0,1,2], inplace=True)
data_L
```

```python
data_L_mean = pd.DataFrame(data_L["N"])
data_L_mean.insert(loc = 1, column='mean', value=data_L["mean"])
data_L_mean.insert(loc = 2, column='faculty', value='ling')


data_M_mean = pd.DataFrame(data_M["N"])
data_M_mean.insert(loc = 1, column='mean', value=data_M["mean"])
data_M_mean.insert(loc = 2, column='faculty', value="match")

data_All_mean = pd.concat([data_L_mean, data_M_mean])
data_All_mean
```

```python
g_L = sns.catplot(data=data_All_mean,
            x='N',
            y='mean',
            kind="bar",
            hue='faculty',
            height=6, aspect=3)

# g_M = sns.catplot(data=data_M_mean,
#             x='N',
#             y='mean',
#             kind="bar",
#             height=6, aspect=3,
#             color='green')

# g_M.set_xticklabels(rotation=75)
g_L.set_xticklabels(rotation=75)
```
